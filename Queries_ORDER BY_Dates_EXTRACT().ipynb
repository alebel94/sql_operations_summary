### ORDER BY

```python
%%sql result << 

SELECT customer_id, amount
FROM payment
WHERE amount >= 6.99
ORDER BY customer_id;
```

```python
# Query for data that shows customers who paid
# an amount GREATER than or EQUAL to $6.99
# The ORDER BY function oredrs the list by customer_id column
# The default is ascending order and it is usually the last line of the query

result
```

```python
%%sql result << 

SELECT customer_id, amount
FROM payment
WHERE amount >= 6.99
ORDER BY amount ASC;
```

```python
# Same as the last example but now it is ordered by amount
# Doing so and specifying in Ascending order (ASC)

result
```

```python
%%sql result << 

SELECT customer_id, amount
FROM payment
WHERE amount <> 11.99
ORDER BY amount DESC;
```

```python
# Query for data that shows customers who paid an amount NOT EQUAL to $0.00
# order the result IN DESCENDING order


result
```
